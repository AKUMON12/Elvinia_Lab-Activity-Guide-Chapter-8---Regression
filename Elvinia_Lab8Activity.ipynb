```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a synthetic dataset
data = {
    'Study_Hours': np.random.randint(1, 10, 50),
    'Test_Score': np.random.randint(50, 100, 50)
}
df = pd.DataFrame(data)

# Display the first few rows of the dataset
df.head()
```

```python
# Scatter plot
plt.scatter(df['Study_Hours'], df['Test_Score'])
plt.title('Study Hours vs. Test Score')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.show()

# Correlation
correlation = df['Study_Hours'].corr(df['Test_Score'])
print("Correlation coefficient:", correlation)
```

```python
from sklearn.linear_model import LinearRegression

# Define the independent and dependent variables
X = df[['Study_Hours']]
y = df['Test_Score']

# Create and fit the model
model = LinearRegression()
model.fit(X, y)

# Predict values
y_pred = model.predict(X)

# Plot regression line
plt.scatter(df['Study_Hours'], df['Test_Score'], color='blue')
plt.plot(df['Study_Hours'], y_pred, color='red')
plt.title('Study Hours vs. Test Score with Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.show()

# Print model coefficients
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
from sklearn.metrics import r2_score

# Calculate R-squared
r_squared = r2_score(y, y_pred)
print("R-squared:", r_squared)
```
